# Backgrounder

Perguruan tinggi kedinasan (PTK) merupakan salah satu tujuan siswa SMA dalam melanjutkan pendidikannya. Pemilihan perguruan tinggi kedinasan sebagai pendidikan lanjutan didasari pada keuntungan yang diberikan oleh instansi tersebut, mulai dari biaya pendidikan yang gratis sampai jaminan menjadi Aparatur Sipil Negara (ASN) ketika lulus nanti. Salah satu perguruan tinggi kedinasan yang menawarkan keuntungan tersebut adalah Politeknik Statistika STIS.

Politeknik Statistika STIS merupakan perguruan tinggi kedinasan di bawah Badan Pusat Statistik (BPS). Berdasarkan rilis Badan Kepegawaian Negara (BKN), jumlah pendaftar di Polstat STIS pada tahun 2023 menduduki urutan ke-3 sebanyak 18.261 orang. Angka ini juga diikuti dengan masifnya jumlah pendaftar di tahun 2023. BKN dalam unggahannya di bulan [April](https://www.instagram.com/p/CrfeGs-LKXH/) menunjukkan jumlah pendaftar di Polstat STIS menduduki peringkat kedua dengan total 12441 pendaftar. Dengan total penerimaan hanya [500 peserta](https://drive.google.com/file/d/19L54hnH4CPuri5q5lvA5U9Qzx_HEvc7u/edit) (keketatan 4.167%), peserta seleksi penerimaan mahasiswa baru haru menganalisis kondisi dan berstrategi agar peluang ia diterima lebih besar.

### Data Tahap 1

Membersihkan dataframe dari data tahap 1. Data tahap 1 SPMB memberikan gambaran mengenai seluruh peserta yang berpartisipasi dalam seleksi masuk Polstat STIS serta peserta yang lolos ke Tahap ke-2.

```python
# Muat data
df_tahap1 = pd.read_csv('Tahap 1.csv')
df_tahap1.head()
```

```python
# Cek jumlah data dan tipe data 
df_tahap1.info()
```

Data dari file tahap 1 berisi 18261 record dengan kondisi yang lengkap/tanpa nilai null. Adapun beberapa variabel perlu disesuaikan tipe datanya, seperti
*   kode_pendidikan
*   kode_jurusan
*   kode_lokasi

Selain itu, atribut 0 juga perlu dihapus karena tidak memberikan makna pada data. 

```python
# Menyesuaikan tipe data
for attr in ['kode_pendidikan', 'kode_jurusan', 'kode_lokasi']:
    df_tahap1[attr] = df_tahap1[attr].astype(object)

# menghapus kolom
df_tahap1.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df_tahap1.columns
```

```python
# mengkonversi nilai pada kolom pendidikan sesuai dengan tingkat pendidikan yang ditempuh peserta
jenis_pendidikan = {
    3000020 : "MADRASAH ALIYAH IPA",
    3004002 : "SMA IPS",
    3001005 : "SMA IPA",
    3000021 : "MADRASAH ALIYAH IPS",
    3171110 : "SMK TEKNOLOGI INFORMASI DAN KOMUNIKASI"
}
df_tahap1['pendidikan'] = df_tahap1['kode_pendidikan'].replace(jenis_pendidikan)

# mendapatkan nilai jenis kelamin berdasarkan nomor peserta
jenkel = {
    "1":"Lk",
    "2":"Pr"
}
df_tahap1['jk'] = df_tahap1['no_peserta'].str[9].replace(jenkel)

# sesuaikan nama kolom dengan penamaan yang diterapkan pada data 2,3,dan 4.
nama_kolom = {
    'no_peserta':'no.peserta',
    'kode_pendidikan':'kode.pendidikan',
    'total_twk':'twk.nilai',
    'total_tiu':'tiu.nilai',
    'total_tkp':'tkp.nilai',
    'total_skd':'skd.nilai',
    'keterangan':'skd.keterangan',
    'kode_jurusan':'kode.prodi',
    'jurusan':'nama.prodi',
    'kode_lokasi':'kode.formasi',
    'lokasi_formasi':'lokasi.formasi',
    'pendidikan':'nama.pendidikan'
}
df_tahap1.rename(columns=nama_kolom, inplace=True)

print(df_tahap1.info())
df_tahap1.head()
```

### Data Tahap 2

Data ini menggambarkan daftar peserta yang lolos ke Tahap ke-3 karena telah memenuhi kriteria pemeringkatan dari tes Matematika dan Psikologi. Data 2, 3, dan 4 diperoleh dari hasil ekstraksi file PDF yang tersedia di [sini](https://www.kaggle.com/datasets/rezkyyayang/spmbpolstatstis2022)

```python
# Memuat data
df_tahap2 = pd.read_csv("Tahap 2.csv", delimiter=";")

# inspeksi data
print(df_tahap2.info())
df_tahap2.head()
```

Pada pengecekan di atas, masih terdapat atribut yang memiliki data yang tidak sesuai. Maka dari itu, akan dilakukan penyesuaian dengan mengubah nilai numberik menjadi objek.

```python
# Menyesuaikan tipe data
for attr in ['kode.formasi', 'kode.prodi', 'no.ujian']:
    df_tahap2[attr] = df_tahap2[attr].astype(object)

# Tipe data mtk.nilai harusnya integer
# namun karena ada peserta yang tidak hadir dan nilainya menjadi "TH", maka kita ubah ini menjadi 0
df_tahap2['mtk.nilai'] = df_tahap2['mtk.nilai'].replace("TH",0)
df_tahap2['mtk.nilai'] = df_tahap2['mtk.nilai'].astype(int)

# cek hasil penyesuaian
print(df_tahap2.info())
```

### Data Tahap 3 dan 4

Data 3 berisi data peserta SPMB yang berhasil lulus dari tes kesehatan dan kebugaran sehingga diterima di Polstat STIS. Di samping data 3, terdapat data 4 yang bertindak sebagai pelengkap kurangnya peserta yang memenuhi kriteria pada daerah tertentu.

Karena tujuan awal dari proses analisis adalah mengetahui faktor yang mempengaruhi kelulusan, maka data 3 dan 4 akan digabung sehingga dihasilkan daftar nama yang lulus saja.

```python
# muat data 3
df_tahap3 = pd.read_csv("Tahap 3.csv", delimiter=";")

# inspeksi data
print(df_tahap3.info())
df_tahap3.head()
```

```python
# sesuaikan tipe data pada data 3
for attr in ['kode.formasi', 'kode.prodi', 'no.ujian', 'peringkat.cadangan']:
    df_tahap3[attr] = df_tahap3[attr].astype(object)

# cek hasil penyesuaian
print(df_tahap3.info())
```

```python
# muat data 4, yang lulus saja
df_tahap4 = pd.read_csv("Tahap 4 Cadangan.csv", delimiter=";")

# inspeksi data
print(df_tahap4.info())
df_tahap4.head()
```

```python
# sesuaikan tipe data pada data 4
for attr in ['kode.formasi.awal', 'kode.formasi.akhir', 'kode.prodi', 'no.ujian', 'peringkat.cadangan']:
    df_tahap4[attr] = df_tahap4[attr].astype(object)

# cek hasil penyesuaian
print(df_tahap4.info())
```

Selain itu, pada hasil akhir di tabel 4 ini terdapat kesalahan data. Setelah ditelisik, ternyata nilai yang benar adalah "Tidak Lulus"

```python
df_tahap4['hasil.akhir'].value_counts()
```

```python
df_tahap4['hasil.akhir'] = df_tahap4['hasil.akhir'].replace("Tidak Lulus Lulus", "Tidak Lulus")
```

```python
# menggabungkan data 3 dan 4 untuk mendapatkan peserta yang lulus
df_lulus = pd.merge(
    left=df_tahap3,
    right=df_tahap4,
    how="left",
    left_on="no.ujian",
    right_on="no.ujian"
)

df_lulus.head()
# df_tahap4[df_tahap4['hasil.akhir']=='Lulus']
```

```python
# mendapatkan data peserta yang lulus saja
df_lulus = df_lulus[~(df_lulus['hasil.akhir_x']=="Tidak Lulus")]

df_lulus['hasil.akhir_x'] = df_lulus.apply(lambda row: "Lulus" if row['hasil.akhir_y'] == "Lulus" else "Tidak Lulus" if row['hasil.akhir_y'] == "Tidak Lulus" else row['hasil.akhir_x'], axis=1)

# menghapus kolom yang tidak digunakan lagi
df_lulus.drop(df_lulus.columns[7:], axis=1, inplace=True)
```

Setelah di cek, ternyata pada atribut "Hasil Akhir" masih terdapat data yang bernilai Lulus Cadangan. Setelah diperiksa, ternyata mereka adalah orang-orang yang lulus cadangan di tahap ketiga namun tidak mendaftar untuk formasi cadangan sehingga tidak tercatat di tahap 4. Maka dari itu, statusnya ditetapkan menjadi Tidak Lulus.

```python
df_lulus['hasil.akhir_x'].value_counts()
```

```python
df_lulus['hasil.akhir_x'] = df_lulus.apply(lambda row: "Tidak Lulus" if row['hasil.akhir_x'] == "Lulus Cadangan" else row['hasil.akhir_x'], axis=1)
df_lulus['hasil.akhir_x'].value_counts()
```

```python
# informasi terkait hasil akhir
print(df_lulus.info())
df_lulus.head()
```

#### Jumlah Partisipan dan Kelulusan

Berdasarkan data yang telah diolah, keketatan dari SPMB Polstat STIS mencapai 2.87%, di mana dari 34 yang mendaftar, hanya 1 orang yang diterima.

```python
# Jumlah yang lulus per provinsi
lulus_per_prov = df_eda[df_eda['skd.keterangan']==1].groupby('lokasi.formasi').agg(
    lulus_skd = ('skd.keterangan','sum'),
    lulus_tahap2 = ('t2.keterangan','sum'),
    lulus_akhir = ('hasil.akhir_x','sum'),
).reset_index()

lulus_per_prov['persen_skd'] = round(lulus_per_prov['lulus_akhir']/lulus_per_prov['lulus_skd']*100,2)

print("Persentase penerimaan berdasarkan jumlah peserta lolos SKD:", lulus_per_prov['persen_skd'].mean())
lulus_per_prov.sort_values('persen_skd', ascending=True)
```

Berdasarkan informasi di atas, Jakarta menjadi lokasi dengan keketatan yang paling tinggi ketika dibandingkan dengan jumlah peserta yang lulus SKD di tahap awal. Dari 50 peserta yang lulus SKD, hanya 9 orang yang dapat dialokasikan pada kuota penerimaan.
